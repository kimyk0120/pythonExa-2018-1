```python
#정규표현식
import re

search_target = '''Luke Skywarker 02-123-4567 luke@daum.net
다스베이더 070-9999-9999 darth_vader@gmail.com 서울시 서초구 서초동 서초아파트
princess leia 010 2454-9999 leia@gmail.com'''
#변수 입력
```

```python
re.search('\d+',search_target).group()
#위의 변수에서 숫자가 한번 이상 연속하는 내용 찾아 출력 (d(숫자),+(한번이상 반복))
```

```python
re.findall('\d+-\d+-\d+',search_target)
#연속된 숫자 - 연속된 숫자 - 연속된 숫자 패턴 모두 찾아서 출력
```

```python
re.findall('\w+@\w+[.]\w+',search_target)
#연속된 문자열 @ 연속된 문자열 . 연속된 문자열 패턴 모두 출력 (w-문자열)
```

```python
re.findall('[가-힣]+', search_target)
#한글이 연속된 패턴 list형으로 모두 출력
```

```python
re.findall('[가-힣]+시\s[가-힣]+구\s[가-힣]+동', search_target)
#연속된 한글 '시' (s- 띄어쓰기) 연속된 한글 '구' (띄어쓰기) 연속된 한글 '동' 
```

```python
tmp = soup.find('dd','txt_desc').get_text()
#soup 내용중 dd태그 txt_desc 속성값을 텍스트 형태로 tmp변수에 저장
result = re.search("\d+.(\d+)?", tmp)
#tmp변수 중 연속된 문자열 . 연속된 문자열이 있거나 없는 패턴 찾아 result변수에 저장
result.group()
```

```python
point = []
#비어있는 point 변수 생성
for each in soup.find_all('dd','txt_desc'):
#soup변수 내용중 dd태그 txt_desc속성값에 해당하는 모든 내용을 each변수에 입력하는 반복문
    result = re.search("\d+.(\d+)?", each.get_text())
#each변수 중 연속된 문자열 . 연속된 문자열이 있거나 없는 패턴 찾아 result변수에 저장
    point.append(result.group())
#result결과는 하나씩 point 변수에 저장
point
```

```python
soup.find('dd','txt_block').get_text()
# soup변수 중 dd태그, txt_block속성값을 갖는 첫번째 내용을 텍스트형태로 출력
```

```python
tmp = soup.find('dd','txt_block').get_text()
# soup변수 중 dd태그, txt_block속성값을 갖는 첫번째 내용을 텍스트형태로 tmp변수에 저장
result = re.search("\d+.\d+.\d+", tmp)
#tmp변수 중 연속된 문자열 . 연속된 문자열 . 연속된 문자열 패턴 찾아 result변수에 저장
result.group()
```

```python
date = []
#비어있는 list형 date변수 생성 
for each in soup.find_all('dd','txt_block'):
# soup변수 중 dd태그, txt_block속성값의 내용을 each변수에 입력
    result = re.search("\d+.\d+.\d+", each.get_text())
# each변수dml 텍스트 형태 중 연속된 문자열 . 연속된 문자열 . 연속된 문자열 패턴 찾아 result변수에 저장    
    date.append(result.group())
#date 변수에 result결과 하나씩 추가
date
```

```python
import pandas as pd
date = pd.date_range('2018.07.01', periods = 35, freq = 'D')
# pandas모듈의 date_range함수를 이용하여 20180701부터 35일 data생성
date
```

```python
import time
import urllib
from urllib.request import Request
from urllib.request import urlopen
from bs4 import BeautifulSoup
movie_date = []
movie_name = []
movie_point = []

for today in date:
    html = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date={date}'
    response = urlopen(html.format(date = today.strftime('%Y%m%d')))
#today변수값의 형식을 변환하여 url주소에 입력
    soup = BeautifulSoup(response, 'html.parser')
    
    movie_date.extend([today]*len(soup.find_all('td','point')))
# 한 날짜에 50위까지 있기때문에 같은 날짜 50번 반복
    movie_name.extend([each.a.string for each in soup.find_all('div','tit5')])
# soup변수 중 div태그 tit5속성값을 갖는 내용을 each에 입력하고 그 중 a태그에 해당하는
# 문자열 movie_name에 저장
    movie_point.extend([each.string for each in soup.find_all('td','point')])
    
    print(str(today))
    time.sleep(1)
```

```python
# 주소, 가격 탐색

price = []
address = []

for idx, row in df.iterrows(): # index, row값이 증가하며 반복문 실행
    html = urlopen(row['URL'])
    soup_tmp = BeautifulSoup(html, 'lxml')
    
    gettings = soup_tmp.find('p','addy').get_text()

    price_tmp = re.split('.,', gettings)[0]
    tmp = re.search('\$\d+\.(\d+)?', price_tmp).group()
    price.append(tmp)

    end = re.search('\$\d+\.(\d+)?', price_tmp).end()
    address.append(price_tmp[end+1:])
    
    time.sleep(0.5)
    print(row['Rank'], '/', '50')
```

```python
df['Price'] = price
df['Address'] = address
# Price, Address열 추가

df = df.loc[:, ['Rank','cafe','Menu','Price','Address']]
#df변수는 전체행과 Rank, cafe, Menu, Price, Address 열로 dataframe 구성
df.set_index('Rank', inplace=True)
#df변수의 index는 Rank로 하고 그 내용을 저장
df.head()
```
