## 인구비율만들기

```python
raw_data
```

```python
raw_data.corr()
```

```python
raw_data['외국인비율'] = raw_data['외국인'] / raw_data['인구수']
raw_data['고령비율'] = raw_data['고령자'] / raw_data['인구수']
raw_data['여성비율'] = raw_data['여성'] / raw_data['인구수']
raw_data.head()
```

```python
raw_data.sort_values(by='인구수', ascending=False)
```

```python
raw_data.sort_values(by='여성비율', ascending=False)
```

```python
raw_data.sort_values(by='고령비율', ascending=False)
```

```python
raw_data.sort_values(by='외국인비율', ascending=False)
```

```python
raw_data.sort_values(by='세대당인구', ascending=True)
```
