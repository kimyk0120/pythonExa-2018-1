```python
import requests
from pandas import *
import datetime

divid = 'adongCd' # 논현동
key = '1168064000'
dtype = 'json'
servicekey = 'XGjk3lw3HWVPQUnt%2FbK0ps7Oun48a%2FVgEb0n1PnbqqERTRXj2wwdybeabsJBVinRb2braJKLp9y0jGc8RpQAdw%3D%3D'
num = '1000'
page = '1'
```

```python
name = [] # 상호명 적재
while True:
    url = 'http://apis.data.go.kr/B553077/api/open/sdsc/storeListInDong?divId={}&key={}&ServiceKey={}&type={}&numOfRows={}&pageNo={}'.format(
        divid, key, servicekey, dtype, num, page)
    r = requests.get(url)
    if r.status_code == 200: # response 제대로 받았는지
#        print(r.text)
        data = r.json()
        for item in data['body']['items']: # Json접근. dict형이라 ky값으로 접근
            name.append(item['bizesNm'])
        if len(data['body']['items']) == 1000:
            print('1000개')
        else:
            break
    page = str(int(page) + 1)
    print(page)

dt = DataFrame({'상호명' : name})
```
