```python
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException, WebDriverException
from bs4 import BeautifulSoup
from pandas import *
import time
import sys
import re
from ckonlpy.tag import Twitter
# pip install customized_konlpy
import matplotlib.pyplot as plt
from collections import Counter
# %matplotlib   
#쥬피터에서 새창에 그래프 그리도록
```

```python
#검색연산자 이용하면 제한적 크롤링 가능
keyword = '손흥민'
dr_ = webdriver.Chrome('/Users/ahn/chromedriver')

#1. 네이버 뉴스 크롤링을 위한 환경 조건설정
url = 'https://search.naver.com/search.naver?where=news&sm=tab_jum&query='+keyword+'&nso=' # 접근 url
dr_.get(url) # url페이지 연결
dr_.find_element_by_xpath('//*[@id="_search_option_btn"]').click() # 검색옵션 선택
dr_.find_element_by_xpath('//*[@id="news_popup"]/a').click() # 언론사 선택
time.sleep(1)
dr_.find_element_by_xpath('//*[@id="ca_p1"]').click() #일간지 선택
time.sleep(1)
dr_.find_element_by_css_selector('button[class = "impact _submit_btn"]').click()  # 확인버튼
try:
    dr_.find_element_by_xpath('//*[@id="main_pack"]/div[2]/div/div[3]/ul/li[2]/a').click() 
except NoSuchElementException:
    dr_.find_element_by_xpath('//*[@id="main_pack"]/div[1]/div/div[3]/ul/li[2]/a').click()
dr_.find_element_by_xpath('//*[@id="nx_option_mson"]').click() # 옵션 유지 버튼 클릭
begin_date = '2018.08.15 '
end_date = '2018.08.28 '
dr_.find_element_by_xpath('//*[@id="snb"]/div/ul/li[2]/a').click() # 기간 탭 선택
time.sleep(5)
dr_.execute_script("document.getElementById('news_input_period_begin').value='{0}'".format(begin_date)) # 날짜입력
time.sleep(5)
dr_.execute_script("document.getElementById('news_input_period_end').value='{0}'".format(end_date)) # 날짜입력
dr_.find_element_by_xpath('//*[@id="_nx_option_date"]/div[2]/span/button').click()

dr_.find_element_by_xpath('//*[@id="snb"]/div/ul/li[2]/a').click()
dr_.find_element_by_xpath('//*[@id="_nx_option_date"]/div[2]/span/button').click()
title_l = []
date_l = []
gisa_l = []

#2. 네이버 뉴스 크롤링
q = True
while q:
    nns_ = dr_.find_elements_by_xpath("//a[contains(@href,'https://news.naver.com/main/read.nhn')]")
    # print(nns_)
    for i in nns_:
        i.click()
        time.sleep(1)
        dr_.switch_to.window(dr_.window_handles[1])
        drt_ = dr_.page_source
        soup_ = BeautifulSoup(drt_,'html.parser')
        try:
            title_ = soup_.find('h3',attrs = {'id':'articleTitle'}).text
        except AttributeError:
            title_ = soup_.find('h2', attrs = {'class' : 'end_tit'}).text
        try:
            gisa_ = soup_.find('div',attrs = {'id':'articleBodyContents'}).text
        except AttributeError:
            gisa_ = soup_.find('div', attrs = {'id' : 'articeBody'}).text
        try:
            date_ = soup_.find('span',attrs = {'class':'t11'}).text
        except AttributeError:
            date_ = soup_.find('span',attrs = {'class' : 'bar'}).text
        gisa_l.append(gisa_[55:])
        # print(gisa_)
        date_l.append(date_)
        title_l.append(title_)
        dr_.close()
        dr_.switch_to.window(dr_.window_handles[0])
        time.sleep(1)
    try:
        dr_.find_element_by_class_name('next').click()
    except NoSuchElementException:
        q = False
print(len(gisa_l),len(date_l),len(title_l))
dt = DataFrame({'gisa' : gisa_l,
                'date' : date_l,
                'title' : title_l})
daterange = begin_date+end_date

dt.to_csv(keyword+'.csv',
                      index=False)
```

```python
## 2. 형태소 분리시행
tw = Twitter()
tw.add_dictionary(['골 손흥민'], 'Noun') # 검색 사전에 없는 어휘 추가, 해당 어휘 추출 가능

dt = read_csv('C:/Users/ahn/'+keyword + '.csv',
             engine = 'python')  #크롤링 프로젝트가 있는 위치에서 검색
gisas = dt['gisa']  # 기사라는 열을 불러오기
gija = re.compile('[가-힣]+ 기자')  #re.compile을 특정 패턴에 맞는 문자 찾아내기. 한글 전체를 찾고, 기자로 끝나는 문장을 찾기.
```

```python
gisapre_ = []
# 전체 돌릴때
for gisa in gisas:
    if re.search(gija, gisa) != None:
        pure_gisa = gisa[:re.search(gija, gisa).span()[0]]
        gisapre_.append(pure_gisa)
    else:
        gisapre_.append(gisa)
        #None이 아니면, gisa에서 giza의 위치를 찾아서 pure_gisa에 저장
        #gisa_pre에 .append(넣어라).
        #none이면 넣고 none이 아니면, 기자를 뺀 나머지를 넣어라.
        #기자가 있는 분면이면, 기자앞을 추출하고, 없으면 전체 추출하는 프로세스
whole_ = []  #전체를 담는 그릇
nounsent_ = []  # 문장단위로 쪼개서 담는 그릇
for pgisa_ in gisapre_:  # 기자제외한 기사 전체에서 기사 하나하나씩을 pgisa에 저장하고자 함.
    sents_ = pgisa_.split('. ')  #점을 기준으로 나누기
    sent_nouns = []
    for sent_ in sents_:  #점을 기준으로 명사인것을 추출하고자 하는데,
        nouns_ = tw.nouns(sent_)  #tw.nouns 명사 추출명령어
        whole_.extend(nouns_)  #명사를 whole_(전체를 담는 그릇)에 넣기 (.extend)
        for noun in nouns_:
            if len(noun) >= 2:
                sent_nouns.append(noun)
    nounsent_.append(sent_nouns)
whole_2 = []
for i in whole_:
    if len(i) >= 2:
        whole_2.append(i)
        #음절 2개 이상 추출
word = list(Counter(whole_2).keys())  # 전체를 카운팅해서 리스트로 만들어서 key(주요단어)와 value(빈도)로 구분
count = list(Counter(whole_2).values())  #빈도는 count_에 넣기
```

```python
# 필요없는 키워드 삭제 함수
def del_many(del_words, word, count):
    if del_words == []:
        pass
    else:
        for i in del_words:
            try:
                index = word.index(i)
                word.remove(word[index])
                count.remove(count[index])
            except ValueError:
                continue
    return [word, count]
```

```python
# 아래 리스트에 넣은것과 같이 필요없는 단어를 추가해서 넣기
del_words = ['사용자','언급','개발','통해']  
del_word_count = del_many(del_words, word, count)
```

```python
del_word_count
```

```python
# 불필요 단어 제외
del_dt = DataFrame({'word' : del_word_count[0],
                   'count' : del_word_count[1]})
del_dt_sort = del_dt.sort_values('count',ascending = False)
```
